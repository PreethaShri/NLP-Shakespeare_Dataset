```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use("seaborn-v0_8")
```

```python
def generatedata(m):
  X= np.random.randn(m)*10
  noise = np.random.randn(m)
  Y = 3*X + 1 + 5*noise
  return X,Y
```

```python
X,Y = generatedata(100)
print(X.shape,Y.shape)
```

```python
def plotdata(X,Y,colors ='green',title='Plotdata'):
  plt.title(title)
  plt.xlabel('X')
  plt.ylabel('Y')
  plt.scatter(X,Y,color=colors)
  plt.show()

plotdata(X,Y)
```

```python
def norma(X):
  X= (X-X.mean())/X.std()
  return X
```

```python
plotdata(X,Y)
X= norma(X)
plotdata(X,Y)
```

```python
m = X.shape[0]
print(m)
```

```python
def traintestsplit(X,Y,split =0.8):
  m = X.shape[0]
  data = np.zeros((m,2))
  data[:,0] = X
  data[:,1] = Y
  np.random.shuffle(data)

  split = int(m*split)

  XT = data[:split,0]
  yT= data[:split,1]
  Xt = data[split:,0]
  yt = data[split:,1]

  return XT,yT,Xt,yt
```

```python
XT,yT,Xt,yt = traintestsplit(X,Y)
```

```python
print(XT.shape,yT.shape,Xt.shape,yt.shape)
```

```python
plt.scatter(XT,yT,color='green',label ="Train")
plt.scatter(Xt,yt,color='red',label ="Test")
plt.title("Train-Test Data")
plt.legend()
plt.show()
```

```python
def hypo(X,theta):
  return theta[0]+theta[1]*X

def error(X,Y,theta):
  m = X.shape[0]
  e = 0
  for i in range(m):
    Y_i = hypo(X[i],theta)
    e = e + (Y[i] - Y_i)**2
    return e/(2*m)

def gradient(X,y,theta):
  m = X.shape[0]
  grad = np.zeros((2,))

  for i in range(m):
    exp = hypo(X[i],theta) - Y[i]
    grad[0] += exp
    grad[1] += exp*X[i]

  return grad/m

def train(X,Y, lr = 0.1, maxItra = 100):
  theta = np.zeros((2,))

  error_list = []

  for i in range(maxItra):
    grad = gradient(X,Y,theta)
    error_list.append(error(X,Y,theta))
    theta[0] = theta[0] - lr*grad[0]
    theta[1] = theta[1] - lr*grad[1]

  plt.xlabel("Iteration")
  plt.ylabel("Error")
  plt.plot(error_list)
  return theta

```

```python
theta = train(X,Y)
```

```python
theta
```

```python
def predict(X,theta):
  return hypo(X,theta)
```

```python
plt.scatter(XT,yT)
plt.scatter(Xt,yt,color ="Violet")
plt.show()
```

```python
yp = predict(Xt,theta)
Xt.shape
yp.shape
```

```python
plt.scatter(XT,yT,color= "Blue",label = "Train")
plt.scatter(Xt,yt,color ="Violet", label = "Test")
plt.plot(Xt,yp,color ="red",label = "Prediction")
plt.legend()
plt.show()
```

```python
def r2score(y,yp):
  num = np.sum((y-yp)**2)
  denum = np.sum((y-y.mean())**2)
  return 1 - num/denum
```

```python
r2score(yt,yp)
```
